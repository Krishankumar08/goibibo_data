```python
import pandas as pd
```

```python
s1=pd.read_csv(r'D:\crime folder\hotel_senti.csv')
s1
```

```python
s2=[]
for i in range (0,len(s1)):
    score=TextBlob(s1.iloc[i][0])
    score1=score.sentiment[0]
    s2.append(score1)
```

```python
g1=pd.concat([s1,pd.Series(s2)],axis=1)
g1.rename(columns={0:'Sentiments_analysis'},inplace=True)
g1
```
